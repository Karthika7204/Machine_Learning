Seperate the data into inputs and target 

```python
#split the data into data and label
inputs = df.drop([60],axis=1)
target = df[60]
print(input)
print(target)
```

Tranning and Test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(inputs,target,test_size = 0.1,stratify = target,random_state = 0)
print(inputs.shape,X_train.shape,X_test.shape)
```

```python
print(X_train)
print(Y_train)
```

Model Tranning --> LogisticRegression

```python
model = LogisticRegression()
```

```python
#tainning the logisticRegression model with tranning data
model.fit(X_train,Y_train)
```

Model Evaluation

```python
#Accuracy on Tranningdata
X_trainning_prediction = model.predict(X_train)
trainning_data_accuracy = accuracy_score(X_trainning_prediction,Y_train)
```

```python
print("Tranning data Accuracy: ",trainning_data_accuracy)
```

```python
#Accuracy on Testdata
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print("Test data Accuracy: ",test_data_accuracy)
```
