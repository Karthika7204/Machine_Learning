```python
# Import the dependencies
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score
```

### Data Collection

```python
# Loading the datasets
df = pd.read_csv('Datasets\mail_data.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.where((pd.notnull(df)),'')
```

```python
df.head()
```

```python
df.shape
```

Label Encoding

```python
# label spam mail as 0;  ham mail as 1;

df.loc[df['Category'] == 'spam'] = 0
df.loc[df['Category'] == 'ham'] = 1
```

```python
df.head()
```

### Model Training

```python
x = df['Message']
y = df['Category']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=3)
print(x.shape,x_train.shape,x_test.shape)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# transform the text data to feature vectors that can be used as input to the Logistic Regression
feature_extraction = TfidfVectorizer(min_df=1, stop_words='english', lowercase=True)

X_train_features = feature_extraction.fit_transform(x_train)
X_test_features = feature_extraction.transform(x_test)

# convert Y_train and Y_test values as integers
y_train = y_train.astype('int')
y_test = y_test.astype('int')
```

```python
model = LogisticRegression()
model.fit(X_train_features,y_train)
```

```python
# Evaluating the training and testing data
train_data_pred = model.predict(X_train_features)
accuracy = accuracy_score(y_train,train_data_pred)
print("Trainin data accuracy: ",accuracy)

test_data_pred = model.predict(X_test_features)
accuracy = accuracy_score(y_test,test_data_pred)
print()
```
