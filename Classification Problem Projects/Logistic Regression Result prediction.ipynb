## Marks Prediction System

Goal ->

We use Linear Regression modal to predict a student is fail or pass based on Hours of sleeping.

Model : LOGISTIC REGRESSION

Sigmoid Function:

𝑃(𝑦=1∣𝑥) = 1 / 1+𝑒−(c+m𝑥)

If:

P > 0.5 → class 1,
P < 0.5 → class 0

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
#sample data
X = np.array([[1],[2],[5],[3],[4]])
y = np.array([0,0,1,1,1])
```

```python
#modal
modal = LogisticRegression()
modal.fit(X,y)
```

```python
#Probability prediction 
prob = modal.predict_proba(X)[:,-1]
y_pred = modal.predict(X)

print("Predicted Probabilities:", prob)
print("Predicted Classes:", y_pred)
```

```python
plt.scatter(X,y,color='blue',label='Actual')
plt.plot(X,y_pred,color='red',label='Probability')
plt.axhline(y=0.5,color='green',linestyle='--',label='Threshold 0.5')
plt.title('Logistic Regression')
plt.xlabel('Hours Studying')
plt.ylabel('Probability of passing')
plt.show()
```
